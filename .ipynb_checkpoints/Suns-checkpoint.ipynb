```python
import pandas as pd
```

```python
games = pd.read_html("https://www.basketball-reference.com/teams/PHO/2019_games.html")[0]
games.drop(columns=['Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 8', 'Notes'], inplace=True)
games.rename(columns={'Unnamed: 7': 'Result'}, inplace=True)
games.set_index(keys='G', inplace=True)
games.dropna(inplace=True)
games['Date'] = pd.to_datetime(games['Date'] + ' ' + games['Start (ET)'])
games.drop(columns=['Start (ET)'], inplace=True)
games
```
